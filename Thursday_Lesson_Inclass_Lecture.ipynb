## The \__init\__() Method <br>
<p>This method is used in almost every created class, and called only once upon the creation of the class instance. This method will initialize all variables needed for the object.</p>

```python
class Car():
    engine = '4.7L'
    
    def __init__(self, wheels):
        self.wheels = wheels
        
ford = Car(4)
print(ford.wheels)
print(ford.engine)
chevy = Car(6)
print(chevy.wheels)
print(chevy.engine)
```

### Exercise 1 - Turn the shopping cart program from yesterday into an object-oriented program

```python
# Create a class called cart that retains items and has methods to add, remove, and show

class Cart():
    content = {}
    
    def modify(self, item, amt):
        if item in self.content:
            self.content[item] += amt
        else:
            self.content[item] = amt
    
    def add(self, item, amt):
        self.modify(item, amt)
        print(f"Added {amt} {item}(s) to the cart.")
    
    def remove(self, item, amt):
        self.modify(item, -amt)
        print(f"Removed up to {amt} {item}(s) from the cart.")
        if self.content[item] <= 0:
            print(f"All {item}s have been removed from the cart.")
            del self.content[item]
        
    def show(self):
        print(f"Your shopping cart: {self.content}")

def getAmount(oper):
    while True:
        amt = input(f"Amount to {oper}: ")
        if amt.isdigit() and (amt := int(amt)) > 0:
            return amt
        else:
            print("Amount must be an integer greater than 0")
        
def add():
    cart.add(input("Item to add: "), getAmount('add'))

def remove():
    cart.remove(input("Item to remove: "), getAmount('remove'))
        
print("Available commands: add, remove, show")
cart = Cart()
commands = {'add': add, 'remove': remove, 'show': cart.show}
while (cmd := input("Enter command: ").lower()) != 'quit':
    if cmd in commands:
        commands[cmd]()
    else:
        print("Unknown command")
cart.show()
print("Thanks for shopping!")
```

### Exercise 2 - Write a Python class which has two methods get_String and print_String. get_String accept a string from the user and print_String print the string in upper case

```python
class StringHolder():
    def get_String(self):
        self.data = input("Gimme a string: ")
    
    def print_String(self):
        print(self.data.upper())

holder = StringHolder()
holder.get_String()
holder.print_String()
```
